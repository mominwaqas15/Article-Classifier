```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup

chrome_driver_path = r'C:\Users\momin\Desktop\chromedriver-win64\chromedriver.exe'

service = Service(chrome_driver_path)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# Function to scroll down to the end of the page
def scroll_to_end():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

# scrape article information
def scrape_articles(csv_writer):
    # Get page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Extract information from each article
    articles = soup.find_all('article', class_='status-publish')
    for article in articles:
        # Extract article title
        title_element = article.find('h2', class_='blog-entry-title')
        title = title_element.text.strip() if title_element else ''

        # Extract article link
        link_element = article.find('h2', class_='blog-entry-title').find('a')
        link = link_element['href'] if link_element else ''

        # Write the extracted information to the CSV file
        csv_writer.writerow([title, link])

# Load the 1st page
url = 'https://www.datasciencecentral.com/?s=Data+Analysis'
driver.get(url)

# Create and open a CSV file for writing
csv_filename = 'Data_Analysis.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Title', 'Link'])  # Write header

    # Scrape articles for the first 50 pages
    for page_number in range(50):
        time.sleep(2)
        print(f"Scraping articles on page {page_number + 1}")

        # Scroll down to load more articles
        scroll_to_end()

        # Scrape articles on the current page and write to CSV
        scrape_articles(csv_writer)

        # Click on the "Next" button
        next_button = driver.find_element(By.CLASS_NAME, 'next.page-numbers')
        next_button.click()

driver.quit()
```

```python
import csv

# filter out common starting and ending text
def filter_content(text, start_text, end_text):
    start_index = text.find(start_text)
    end_index = text.find(end_text, start_index)
    if start_index != -1 and end_index != -1:
        return text[start_index + len(start_text):end_index].strip()
    return text

# process a CSV file
def process_csv(input_filename, output_filename, start_text, end_text):
    with open(input_filename, 'r', newline='', encoding='utf-8') as infile:
        reader = csv.reader(infile)
        header = next(reader)

        with open(output_filename, 'w', newline='', encoding='utf-8') as outfile:
            writer = csv.writer(outfile)
            writer.writerow(header)  # Write the header

            for row in reader:
                # common content is in 3rd (2) column.
                row[2] = filter_content(row[2], start_text, end_text)
                writer.writerow(row)

input_csv_file = 'Data_Visualization.csv'
output_csv_file = 'Data_Visualization_Cleaned.csv'
common_start_text = 'Data Science Central Data Science Central'
common_end_text = 'Related Content'

process_csv(input_csv_file, output_csv_file, common_start_text, common_end_text)
```

### removing special characters and lowering text case

```python
file_names = ['Machine_Learning_Cleaned.csv', 'Artificial_Intelligence_Cleaned.csv', 'Big_Data_Cleaned.csv', 'Data_Analysis_Cleaned.csv', 'Data_Visualization_Cleaned.csv']

for file_name in file_names:
    df = pd.read_csv(file_name)

    # Data Cleaning
    # Handling NaN values and removing special characters & converting text to lowercase,
    df['Article Content'] = df['Article Content'].fillna('')  # Replace NaN with an empty string
    df['Article Content'] = df['Article Content'].apply(lambda x: ' '.join(str(x).lower() for x in str(x).split()))
    df['Article Content'] = df['Article Content'].str.replace('[^a-zA-Z\s]', '')

    # Save to new CSV file
    cleaned_file_name = f'{file_name}'
    df.to_csv(cleaned_file_name, index=False)

    print(f"Cleaned data saved to {cleaned_file_name}")
```

# Combining all datasets to get one Main Dataset

```python
import os
import pandas as pd

file_names = ['Machine_Learning_Cleaned.csv', 'Artificial_Intelligence_Cleaned.csv', 'Big_Data_Cleaned.csv', 'Data_Analysis_Cleaned.csv', 'Data_Visualization_Cleaned.csv']

dfs = []
for cleaned_file_name in file_names:
    # Extract label from file name
    label = cleaned_file_name.split('_')[0] + " " + cleaned_file_name.split('_')[1]

    # Read the cleaned CSV file into a DataFrame
    df = pd.read_csv(cleaned_file_name)
    
    # Add a 'Label' column with the extracted label
    df['Label'] = label
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into one DataFrame
all_data = pd.concat(dfs, ignore_index=True)

# Save the concatenated data to a new CSV file
concatenated_file_name = 'data.csv'
all_data.to_csv(concatenated_file_name, index=False)

print(f"Concatenated data saved to {concatenated_file_name}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Convert frozenset to a list
custom_stopwords = list(ENGLISH_STOP_WORDS)

df = pd.read_csv('data.csv')
df.dropna(subset=['Article Content', 'Title', 'Label'], inplace=True)

# Combine 'Article Content' and 'Article Title' columns into a single feature
df['Combined_Content'] = df['Article Content'] + ' ' + df['Title']

X = df['Combined_Content']
y = df['Label']

# Reset index to have continuous integer indices
df.reset_index(drop=True, inplace=True)

# Vectorize the text data using TF-IDF with stopwords removal
vectorizer = TfidfVectorizer(ngram_range=(1, 1), stop_words=custom_stopwords)
X_vectorized = vectorizer.fit_transform(X)

# Initialize and train the Naive Bayes model
naive_bayes = MultinomialNB()
naive_bayes.fit(X_vectorized, y)

# Make predictions on the entire dataset
predictions = naive_bayes.predict(X_vectorized)

# Calculate accuracy for each instance
instance_accuracy = [1 if pred == true_label else 0 for pred, true_label in zip(predictions, y)]

# Add instance accuracy as a new column in the DataFrame
df['Instance_Accuracy'] = instance_accuracy

# Group by class and select the top 100 instances based on accuracy
df_subset = df.groupby('Label').apply(lambda x: x.nlargest(100, 'Instance_Accuracy')).reset_index(drop=True)

# Save the selected instances into a new CSV file
df_subset.to_csv('subset.csv', index=False)
```
